## 2-1. 데이터 오브젝트
### (1) 시리즈(Series)

* **DataFrame**은 **2차원 배열**의 형태를 가짐

## 2-2. Index와 Column
* 데이터 프레임의 기본 구조는 '행(Index)'과 열(Column)'으로 설명 가능 //행: row와 index가 약간 다른 개념인가?

* **컬럼**: 데이터의 카테고리(어떤 자료인지)를 나타냄

## 2-3. 시리즈(Series)

### (1) Series란?
* 값(value)과 인덱스(index)로 구성된 가장 간단한 1차원 자료구조
    * 인덱스가 있는 1차원 배열(array)
    * Series를 구성하는 값들의 데이터 타입은 모두 같아야 함. 
* 인덱스 레이블 지정
    * 별도의 인덱스 레이블을 지정하지 않으면, 자동으로 0부터 시작되는 default 정수형 인덱스가 사용됨
* DataFrame에서 특정 컬럼 1개만 추출한다면, 해당 구조는 Series라고 볼 수 있음.

### (2) Series 만드는 방법
1. 원하는 값들이 포함된 리스트를 생성한다.
2. pd.Series() 함수를 사용하여 Series 객체를 생성한다.
    * 이때, 인덱스를 직접 지정해주고 싶으면 pd.Series() 함수의 인자에 인덱스를 추가한다.
---
```python
import pandas as pd
# 1. 리스트 생성
series = ['LG', '애플', '애플', '삼성', 'ASUS', '삼성']

# 2-1. Series 객체 생성 (인덱스 지정 x)
sr = pd.Series(series)
print(sr)
print()

# 2-2. Series 객체 생성 (인덱스 직접 지정)
index = ['a', 'b', 'c', 'd', 'e', 'f']
sr = pd.Series(series, index = index)
print(sr)
```

## 2-4. 데이터프레임(DataFrame)
### (1) DataFrame이란?
* 행(index)과 열(column)으로 이루어진 2차원 배열 구조
* 길이가 같은 Series들의 집합체로 볼 수 있음
* Series 단위로는 값들의 데이터 타입이 동일해야 함
    * 하나의 컬럼의 값들의 데이터 타입은 동일해야 함
    * 컬럼 간의 데이터 타입은 달라도 됨
### (2) DataFrame 만드는 방법
1. 사전(Dictionary) 생성: 각 카테고리별 네이밍(key), 데이터(value)
2. pd.DataFrame() 함수를 사용하여 데이터프레임 구성

# 3. 데이터 추출
* 앞으로 보게 될 데이터의 양이 매우 방대하므로, 효율적인 분석을 위해 관심 있는 부분만 따로 뽑아내어 가공하는 스킬이 중요함.

## 3-1. 인덱싱(Indexing)과 슬라이싱(Slicing)
### (1) 인덱싱(Indexing)
* 입력하는 조건이 범위가 아니라 **단순 지목**일 경우
### (2) 슬라이싱(Slicing)
* 입력하는 조건이 **범위**인 경우

## 3-2. 행단위 데이터 추출: loc, iloc
### (1) loc
* 기본적으로는 행단위 데이터 추출하는 메서드
* 인자 지정 방법에 따라 열단위/특정 값만 추출할 수도 있음
* 인덱싱&슬라이싱 모두 가능
* **loc 인자 안**에는 **행과 열의 이름(label)** 이 들어와야 함
    * **df.loc[행 이름]** => 행단위 추출 (인덱스 이름이 '행 이름'인 데이터를 추출)
    * **df.loc[행 이름, 열 이름]** => 값단위 추출
    * **df.loc[:, 열 이름]** => 열단위 추출 (전체 인덱스 중에 컬럼 이름이 '열 이름'인 데이터를 추출)
* **(참고) Fancy 인덱싱**
    * 기본적으로 인덱싱은 차원을 하나 축소시키고, 슬라이싱은 차원을 바꾸지 않음
        * df.loc[행 이름, 열 이름]에서 행과 열을 모두 인덱싱했을 때 단일값이 나옴
          : 인덱싱을 2번 해서 차원이 2번 축소되었기 때문
    * Fancy 인덱싱은 **차원을 그대로 유지하면서 인덱싱**하는 방법
    * 여러 개의 값을 순서와 상관없이 선택하고 싶을 때 주로 사용
    * 범위 지정X, 지목 방식 -> 인덱싱과 유사
    * 여러 개의 값을 추출(차원 변화 없음) -> 슬라이싱과 유사
        * 예시
        * df.loc[[0,5],['노트북 브랜드', '사용자 이름']]
        * 인덱스가 0과 5고, 컬럼이 '노트북 브랜드'와 '사용자 이름'인 데이터 추출
### (2) iloc
* loc과 거의 유사한 메서드
* **iloc 인자 안**에는 **정수형 인덱스 번호**가 와야 함
* 데이터의 양이 많아질 경우, iloc이 속도 측면에서 더 유리함.

    * **df.iloc[행 순서, 열 순서]**
    * **(참고) df.iloc의 범위: (시작 인덱스 ~ 끝 인덱스 -1)**
        * df.loc[0:2] # 끝 인덱스를 포함하여 반환 (0, 1, 2 행 반환)
        * df.iloc[0:2] # 끝 인덱스를 포함하지 않고 반환 (0, 1 행 반환)

## 3-3. 열단위 데이터 추출: 컬럼명
* loc, iloc으로도 열단위 데이터 추출이 가능하긴 함
* 하지만 열단위로만 데이터를 추출하고자 한다면 컬럼명으로 인덱싱하는 것이 더 직관적이고 좋은 방식임
    * **df[열 이름]** # 하나의 열만 인덱싱하면 Series를 반환함
    * **df[[열 이름1, 열 이름2, ...]]** # 여러 개의 열을 인덱싱하면 DataFrame의 부분을 반환함
    * **df[[열 이름]]** # 하나의 열만 반환하면서 DataFrame 자료형을 유지함 (Fancy?)
    * **df[열 이름][행 이름]** # '열 이름' 컬럼에서 '행 이름' 인덱스의 행을 반환함

## 3-4. Boolean Indexing
### (1) 불리언 인덱싱(Boolean Indexing)이란?
* 인덱스/컬럼명이 아니라, **특정 조건을 만족하는 데이터를 추출**하고 싶을 때 사용하는 기법
* **True 또는 False 배열**을 통해 **데이터를 필터링**하는 연산
* pandas에서 DataFrame의 특정 열에 조건문을 걸면, 각 값들에 대응하는 True/False 배열이 반환됨
* 앞서 행단위/열단위 데이터 추출에서 인덱스/컬럼명을 넣었던 자리에 **조건문을 넣으면** True 값에 해당하는 부분만 출력됨
### (2) 불리언 인덱싱의 활용
* 조건문 여러 개를 **논리연산자**로 묶어 사용할 수 있음
    * 단, **and는 &** 로, **or은 |** 로 작성해야 오류가 발생하지 않음.
* **loc**에서 불리언 인덱싱 사용 **가능**
* **iloc**에서는 불리언 인덱싱 사용 **불가능**

* **df[df['노트북 브랜드']=='애플']**
    * 노트북 브랜드가 '애플'인 행들만 인덱싱
* **df[(df['노트북 브랜드']=='애플')&(df['노트북 무게(kg)']<1.6)]**
    * 두 조건문을 and(&) 연산자로 함께 사용
    * 노트북 브랜드가 '애플'이면서 노트북 무게가 1.6kg 미만인 행들만 인덱싱
* **df.loc[df['노트북 화면 크기(inch)']>=14, '노트북 무게(kg)']**
    * loc에서도 boolean indexing 사용 가능
    * 노트북 화면 크기가 14인치 이상인 행들의 노트북 무게 컬럼을 추출함

## 3-5. 인덱스 설정
* default 인덱스: 0부터 시작하는 정수형 인덱스
* 인덱스는 원하는 대로 변경 가능
### (1) 기존의 열 중 하나를 인덱스로 설정: df.set_index(열 이름)
* **df.set_index('사용자 이름', inplace=True)**
    * 기존의 열(사용자 이름)을 인덱스로 설정함
    * inplace = True를 추가하면 원본 데이터인 df에 직접적으로 수정이 가해짐
### (2) 인덱스를 리셋하고 기존 인덱스를 열에 추가: df.reset_index()
* **df.reset_index(inplace = True, drop = False)**
    * 인덱스로 설정한 열(사용자 이름)을 일반 열로 되돌리고, 인덱스를 디폴트 정수형 인덱스로 리셋함
    * drop = True를 추가하면 기존 인덱스가 열에 추가되지 않고 사라짐.
    * 디폴트값은 False
### (3) 인덱스의 재배열
* **df.reindex([6,5,4,3,2,1,0])**
    * 인덱스의 재배열과 함께 기존 행 데이터들도 함께 이동함
    * 0번은 연동운, 6번은 김현서
* **df.reindex([0,1,2,3,'a','b','c'])**
    * 기존에 없던 인덱스가 추가되면 값들에 결측치가 기록됨
    * NaN 값이 채워짐
    * 0, 1, 2, 3에 있던 데이터들은 남고, 4, 5, 6에 있던 데이터들은 없어짐
    * 'a', 'b', 'c' 행에는 NaN값이 채워짐
* **df.index = [6,5,4,3,2,1,0]**
    * 인덱스를 재설정하고 싶지만, 데이터들은 제자리에 두고 싶을 때 사용
    * 0번은 김현서, 6번은 연동운

# 4. 데이터 조작
## 4-1. 데이터 추가 및 삭제
### (1) 데이터 추가
* 특정 데이터 셋을 열로 추가할 수 있음
* 특정 리스트를 데이터 프레임에 추가
    * **df['새로운 열 이름'] = 리스트**
    * **df['화면 크기 대비 무게'] = df['노트북 무게(kg)']/df['노트북 화면 크기(inch)']**
        * '화면 크기 대비 무게'라는 새로운 열 추가
### (2) 데이터 삭제
* 특정 열을 삭제할 수 있음
    * **df.drop(labels, axis)**
    * **df.drop('화면 크기 대비 무게', axis=1, inplace=True)**
        * axis=1 : 입력된 label을 열방향으로 탐색해 해당 열을 삭제함
        * inplace = True 옵션 (파괴적 방식) 사용 : 원본 df에 바로 적용됨
    * **df.drop([0,1])**
        * axis=0 (default): 입력된 label을 행방향으로 탐색해 해당 행을 삭제함
        * inplace = False (default, 비파괴적 방식) : 원본 df에 바로 적용되지 않음
        * 한번에 여러 값을 삭제하고 싶은 경우, 리스트를 씌워서 인자로 전달하면 됨.

## 4-2. 데이터 정렬
### (1) 인덱스를 기준으로 정렬: df.sort_index
* **df.sort_index(ascending=False)**
    * ascending=False 옵션: 내림차순으로 정렬 (4,3,2,1...)
    * ascending=True 가 디폴트 (오름차순) (1,2,3,4...)

### (2) 특정 열의 값을 기준으로 정렬: df.sort_values(by="열 이름")
* **df.sort_values(by='노트북 무게(kg)')**
    * 컬럼 기준으로 정렬됨.
    * ascending=True 가 디폴트 (오름차순) (1,2,3,4...)

## 4-3. 데이터 변환
### (1) 특정 컬럼 내 데이터의 자료형 변환
* **df.dtypes**
    * (참고) pandas에서 object는 문자열, 복잡한 형태의 숫자 자료형을 의미함.
* **df[열 이름] = df[열 이름].astype(바꿀 자료형)**
    * **df['노트북 화면 크기(inch)']=df['노트북 화면 크기(inch)'].astype('str')**
    * 비파괴적 방식으로 자료형 바꾸는 방법
    * astype에는 inplace 옵션을 사용할 수 없음
    * 원본을 직접 바꾸고 싶은 경우 df의 칼럼을 한 번 더 지정해주어야 함
### (2) 데이터의 값 자체를 변환
* **df.replace(바꾸고 싶은 값, 바꿀 값)**
    * **df.replace('김현서', '기면서')**
    * inplace=False 가 디폴트 (비파괴적 방식)

## 4-4. 데이터프레임 병합
### (1) concat
* 단순히 두 프레임을 이어붙이는 방식
* **행 방향 결합**: df = pd.concat([df1, df2], axis=0)
* **열 방향 결합**: df = pd.concat([df1, df2], axis=1)

### (2) merge
* 공통 항목을 기준으로 병합하는 방식
* DataFrame 간 공통 항목(열이나 인덱스)이 있어야 합칠 수 있음
    * 형태가 다르더라도, 겹치는 데이터를 기준으로 데이터프레임을 합칠 수 있음
    * 어라? 이거 sql의 조인 중에 뭐랑 좀 비슷한 것 같다..뭐였지?
* **on 인자**를 통해 기준이 될 열을 지정해줄 수 있음
* **집합 연산**과 유사하게 작동
* **how 인자**에 따라 어떤 방식으로 결합할지 결정할 수 있음

* **교집합(and) 연산**: 
    * **df = pd.merge(df1, df2, how='inner', on='~~')**
    * pd.merge(df, df3, how='inner', on='사용자 이름')
        * '사용자 이름' 열을 기준으로 교집합(and) 연산
        * 사용자 이름이 완전히 겹치는 것만 출력
* **합집합(or) 연산**:
    * **df = pd.merge(df1, df2, how='outer', on='~~')**
    * pd.merge(df, df3, how='outer', on='사용자 이름')
        * '사용자 이름' 열을 기준으로 합집합(or) 연산
        * 어느 한쪽에라도 값이 있다면 전부 출력
        * 겹치지 않는 부분은 NaN 값으로 반환

# 5. 집계함수
## 5-1. 집계함수란?
* **여러 행으로부터 하나의 결과값을 반환하는 함수**
* 결측치가 있을 경우 결측치는 무시하고 계산함

## 5-2. 대표적인 집계함수
* 평균: df['컬럼명'].mean()
* 중앙값: df['컬럼명'].median()
* 표준편차: df['컬럼명'].std()
* 합계: df['컬럼명'].sum()
* 최대값: df['컬럼명'].max()
* 최소값: df['컬럼명'].min()

## 5-3. 집계함수의 축 방향
* 각각의 열들에 대한 결과값을 얻고 싶다면 행방향 연산(axis=0)
* 각각의 행들에 대한 결과값을 얻고 싶다면 열방향 연산(axis=1)

## 5-4. 그룹화(Grouping)
* 일부 행을 그룹화하여 집계할 경우 사용
* **groupby** 메서드: 특정 기준으로 그룹을 묶고, 연산처리를 하게 해줌
* groupby 자체만으로는 아무 일도 일어나지 않기 때문에 보통 **집계함수와 함께 사용**함
* **df.groupby('컬럼명1')['컬럼명2'].집계함수**
    * 컬럼명1로 그룹화를 하고, 각각의 그룹별로 컬럼명2에 대해 집계함수를 적용
    * df.groupby('노트북 브랜드')[['노트북 무게(kg)']].mean()
        * 노트북 브랜드 별로 묶어 노트북 무게의 평균을 확인
