```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

SEED = 42 # random state ÌÜµÏùº
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler, LabelEncoder

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.ensemble import (
    VotingClassifier,
    RandomForestClassifier, AdaBoostClassifier,
    GradientBoostingClassifier, StackingClassifier
)
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostClassifier
```

### Îç∞Ïù¥ÌÑ∞ Î°úÎìú

```python
# dfÎùºÎäî Î≥ÄÏàòÏóê 'Employee.csv' ÌååÏùºÏùÑ ÏùΩÏñ¥ÏôÄ Ï†ÄÏû•Ìï©ÎãàÎã§.
df = pd.read_csv('Employee.csv')
```

```python
df.head()
```

```python
target = 'LeaveOrNot'
```

```python
X = df.drop(columns=['LeaveOrNot']) # target Ïª¨ÎüºÏùÑ Ï†úÏô∏Ìïú Îç∞Ïù¥ÌÑ∞Î•º XÏóê Ï†ÄÏû•Ìï©ÎãàÎã§.
y = df[target]
```

### Îç∞Ïù¥ÌÑ∞ Ï†ÑÏ≤òÎ¶¨

```python
X.info()
```

```python
X.head()
```

```python
# Î≤îÏ£ºÌòï Î≥ÄÏàò Ï†ÑÏ≤òÎ¶¨
# ÏùºÍ¥ÑÏ†ÅÏúºÎ°ú Ïù∏ÏΩîÎî©ÏùÑ ÏßÑÌñâÌïòÍ≤†ÏäµÎãàÎã§.

cat_cols = [
    'Education',
    'City',
    'Gender',
    'EverBenched'
]

for col in cat_cols:
    le = LabelEncoder()

    # Í∞Å Î≤îÏ£ºÌòï Î≥ÄÏàò(col)Ïóê ÎåÄÌï¥ Ïù∏ÏΩîÎî©ÏùÑ ÏàòÌñâÌï©ÎãàÎã§.
    X[col] = le.fit_transform(X[col].astype(str))
```

```python
# Ïä§ÏºÄÏùºÎßÅ

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### ÌõàÎ†®Ïö© / ÌÖåÏä§Ìä∏Ïö© Îç∞Ïù¥ÌÑ∞ Î∂ÑÌï†

```python
# X_scaledÏôÄ yÎ•º ÌõàÎ†®Ïö©Í≥º ÌÖåÏä§Ìä∏Ïö© Îç∞Ïù¥ÌÑ∞Î°ú Î∂ÑÌï†Ìï©ÎãàÎã§.
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y, 
    random_state=SEED
)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

#### 1. Logistic Regression

```python
# Î™®Îç∏ÏùÑ Î∂àÎü¨ÏôÄ ÌõàÎ†®ÏãúÌÇµÎãàÎã§.
lr = LogisticRegression(random_state=SEED)
lr.fit(X_train, y_train)
```

```python
# Î™®Îç∏ÏùÑ ÌÜµÌï¥ ÏòàÏ∏°ÏùÑ ÏàòÌñâÌï©ÎãàÎã§.
y_lr_pred = lr.predict(X_test)
```

```python
# Í≥ÑÏÇ∞Îêú ÏòàÏ∏°Í∞íÏóê ÎåÄÌï¥ Ï†ïÌôïÎèÑÎ•º Í≥ÑÏÇ∞Ìï©ÎãàÎã§.
lr_accuracy = accuracy_score(y_test, y_lr_pred)
lr_report = classification_report(y_test, y_lr_pred)

print(f"Accuracy score : {lr_accuracy}")
print(f"Report :\n{lr_report}")
```

#### 2. ÏùòÏÇ¨Í≤∞Ï†ïÎÇòÎ¨¥

```python
tree = DecisionTreeClassifier(random_state=SEED)
tree.fit(X_train, y_train)
```

```python
tree_accuracy = accuracy_score(y_test, y_tree_pred)
tree_report = classification_report(y_test, y_tree_pred)

print(f"Accuracy score : {tree_accuracy}")
print(f"Report :\n{tree_report}")
```

#### 3. SVM

```python
svm = SVC(random_state=SEED)
svm.fit(X_train, y_train)
```

```python
y_svm_pred = svm.predict(X_test)
```

```python
svm_accuracy = accuracy_score(y_test, y_svm_pred)
svm_report = classification_report(y_test, y_svm_pred)

print(f"Accuracy score : {svm_accuracy}")
print(f"Report :\n{svm_report}")
```

#### 4. kNN

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
y_knn_pred = knn.predict(X_test)
```

```python
knn_accuracy = accuracy_score(y_test, y_knn_pred)
knn_report = classification_report(y_test, y_knn_pred)

print(f"Accuracy score : {knn_accuracy}")
print(f"Report :\n{knn_report}")
```

### 4. Stacking

`StackingClassifier`

- **estimators**: Í∏∞Î≥∏ Î™®Îç∏ Î™©Î°ù (Ïù¥Î¶Ñ, Î™®Îç∏) ÏåçÏùò Î¶¨Ïä§Ìä∏
- **final_estimator**: ÏµúÏ¢Ö Î∂ÑÎ•òÍ∏∞ (Í∏∞Î≥∏Í∞í: LogisticRegression)
- **cv**: ÍµêÏ∞® Í≤ÄÏ¶ù Ìè¥Îìú Ïàò (Í∏∞Î≥∏Í∞í: 5)
- **passthrough**: ÏõêÎ≥∏ ÌäπÏÑ±ÏùÑ Î©îÌÉÄ Î™®Îç∏Ïóê Ï†ÑÎã¨ Ïó¨Î∂Ä (Í∏∞Î≥∏Í∞í: False)
- **n_jobs**: Î≥ëÎ†¨ Ï≤òÎ¶¨ Ïàò

[Docs](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.StackingClassifier.html)

```python
base_learners = [
    ('lr', LogisticRegression(random_state=SEED)),
    ('dt', DecisionTreeClassifier(random_state=SEED)),
    ('rf', RandomForestClassifier(random_state=SEED))
]

stack_clf = StackingClassifier(
    estimators=base_learners,
    final_estimator=LogisticRegression(random_state=SEED), # ÏµúÏ¢Ö Î∂ÑÎ•òÍ∏∞Î•º LogisticRegression(random_state=SEED) Î°ú ÏßÄÏ†ïÌï¥Ï£ºÏÑ∏Ïöî
    cv=5
)

# Î™®Îç∏ ÌõàÎ†®
stack_clf.fit(X_train, y_train)

# ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÏòàÏ∏° ÏàòÌñâ
y_stk_pred = stack_clf.predict(X_test)

# Ï†ïÌôïÎèÑ Ï∂úÎ†•
print(f"Stacking Accuracy: {accuracy_score(y_test, y_stk_pred):.4f}")
```

- Í∏âÏó¨Í∞Ä ÎÜíÏùÄ Í≤ΩÏö∞Ïù∏ PaymentTierÏùò Í∞íÏù¥ 1Ïù∏ Í≤ΩÏö∞Ïóê ÎåÄÌï¥ Í≥†Í∏âÏó¨ Ïó¨Î∂ÄÎ•º ÌÜµÌï¥ Í∏âÏó¨ Ïó¨Î∂ÄÏóê Îî∞Î•∏ Ïù¥ÏßÅÏú®ÏùÑ Î∂ÑÏÑùÌï† Ïàò ÏûàÏùå
- ÎòêÌïú, EverBenchedÎäî Yes/NoÎ°ú Íµ¨Î∂ÑÎêòÏñ¥ ÏûàÏúºÎØÄÎ°ú, Ïù¥Î•º 1/0ÏúºÎ°ú Î≥ÄÌôòÌï† Ïàò ÏûàÏùå
  

```python
from sklearn.preprocessing import LabelEncoder

# ‚úÖ 1. Î≤îÏ£ºÌòï Î≥ÄÏàò Ïù∏ÏΩîÎî© (Label Encoding)
label_cols = ['Education', 'City', 'Gender']
for col in label_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))

# ‚úÖ 2. ÌååÏÉù Î≥ÄÏàò ÏÉùÏÑ±
current_year = 2025
df['YearsAtCompany'] = current_year - df['JoiningYear']                   # Ï°∞ÏßÅ ÎÇ¥ Ïû¨ÏßÅ Í∏∞Í∞Ñ
df['IsHighTier'] = (df['PaymentTier'] == 1).astype(int)                  # Í≥†Í∏âÏó¨ Ïó¨Î∂Ä

# ‚úÖ 3. Ïù¥ÏßÑ Î≥ÄÏàò Ï≤òÎ¶¨
df['EverBenched'] = df['EverBenched'].map({'Yes': 1, 'No': 0})
```

## 3. ÏïôÏÉÅÎ∏î Í∏∞Î≤ï ÌôúÏö©

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.ensemble import (
    BaggingClassifier, VotingClassifier, RandomForestClassifier,
    AdaBoostClassifier, GradientBoostingClassifier
)
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
import pandas as pd

# üéØ Feature/Target ÏÑ§Ï†ï
X = df.drop(columns=['LeaveOrNot'])
y = df['LeaveOrNot']

# ‚úÖ Stratified K-Fold ÏÑ§Ï†ï
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
results = {}

# ‚úÖ 1. Bagging
bagging = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=50, random_state=42)
results['Bagging'] = cross_val_score(bagging, X, y, cv=cv, scoring='accuracy').mean()

# ‚úÖ 2. Hard Voting
hard_voting = VotingClassifier(estimators=[
    ('lr', LogisticRegression(max_iter=1000)),
    ('dt', DecisionTreeClassifier()),
    ('nb', GaussianNB())
], voting='hard')
results['Hard Voting'] = cross_val_score(hard_voting, X, y, cv=cv, scoring='accuracy').mean()

# ‚úÖ 3. Soft Voting
soft_voting = VotingClassifier(estimators=[
    ('lr', LogisticRegression(max_iter=1000, solver='liblinear')),
    ('rf', RandomForestClassifier(n_estimators=100)),
    ('gb', GradientBoostingClassifier())
], voting='soft')
results['Soft Voting'] = cross_val_score(soft_voting, X, y, cv=cv, scoring='accuracy').mean()

# ‚úÖ 4. Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
results['Random Forest'] = cross_val_score(rf, X, y, cv=cv, scoring='accuracy').mean()

# ‚úÖ 5. AdaBoost
ada = AdaBoostClassifier(n_estimators=100, random_state=42)
results['AdaBoost'] = cross_val_score(ada, X, y, cv=cv, scoring='accuracy').mean()

# ‚úÖ 6. Gradient Boosting (GBM)
gbm = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
results['Gradient Boosting'] = cross_val_score(gbm, X, y, cv=cv, scoring='accuracy').mean()

# ‚úÖ 7. XGBoost
xgb = XGBClassifier(n_estimators=100, use_label_encoder=False, eval_metric='logloss', random_state=42)
results['XGBoost'] = cross_val_score(xgb, X, y, cv=cv, scoring='accuracy').mean()

# ‚úÖ 8. CatBoost
catboost = CatBoostClassifier(iterations=100, verbose=0, random_state=42)
results['CatBoost'] = cross_val_score(catboost, X, y, cv=cv, scoring='accuracy').mean()

# ‚úÖ Í≤∞Í≥º Ï†ïÎ¶¨
results_df = pd.DataFrame(list(results.items()), columns=['Model', 'Mean CV Accuracy'])
results_df.sort_values(by='Mean CV Accuracy', ascending=False, inplace=True)
print(results_df)
```
