### 1. ë°ì´í„° ë¡œë“œ

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
```

MNIST ë°ì´í„°ì…‹ì„ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤. cache=TrueëŠ” ë‹¤ìš´ë¡œë“œí•œ ë°ì´í„°ë¥¼ ë¡œì»¬ ë””ìŠ¤í¬ì— ì €ì¥í•˜ì—¬ ë‹¤ìŒ í˜¸ì¶œ ë•Œ ë¹ ë¥´ê²Œ ë¡œë“œí•©ë‹ˆë‹¤.

```python
mnist = fetch_openml('mnist_784', version=1, cache=True, as_frame=False)
```

```python
mnist.target = mnist.target.astype(np.int8)
```

í•™ìŠµ íš¨ìœ¨ì„ ìœ„í•´ ëª¨ë“  ì´ë¯¸ì§€ ë°ì´í„°ë¥¼ 0-255 ë²”ìœ„ì—ì„œ 0-1 ë²”ìœ„ë¡œ ì •ê·œí™”í•©ë‹ˆë‹¤.\
ë ˆì´ë¸” ë°ì´í„°ë¥¼ yì— ì €ì¥í•©ë‹ˆë‹¤.

```python
X = mnist.data / 255  # 0-255ê°’ì„ [0,1] êµ¬ê°„ìœ¼ë¡œ ì •ê·œí™”
y = mnist.target
```

ì •ê·œí™”ëœ ì²« ë²ˆì§¸ ì´ë¯¸ì§€ ë°ì´í„°ì™€ ë ˆì´ë¸”ì„ ì¶œë ¥í•©ë‹ˆë‹¤.

```python
plt.imshow(X[0].reshape(28, 28), cmap='gray')
print("ì´ ì´ë¯¸ì§€ ë°ì´í„°ì˜ ë ˆì´ë¸”ì€ {:.0f}ì´ë‹¤".format(y[0]))
```

### 2. í›ˆë ¨ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ì˜ ë¶„ë¦¬

MNIST ë°ì´í„°ì…‹ì„ ë¶ˆëŸ¬ì˜¤ê³ , ì´ë¥¼ í›ˆë ¨ ì„¸íŠ¸ì™€ í…ŒìŠ¤íŠ¸ ì„¸íŠ¸ë¡œ ë¶„í• í•©ë‹ˆë‹¤.

```python
import torch
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/7, random_state=42)
```

```python
# í…ì„œë¡œ ë³€í™˜
X_train = torch.Tensor(X_train)
X_test = torch.Tensor(X_test)
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)

# TensorDataset ê°ì²´ ìƒì„±
ds_train = TensorDataset(X_train, y_train)
ds_test = TensorDataset(X_test, y_test)

# DataLoader ê°ì²´ ìƒì„±
loader_train = DataLoader(ds_train, batch_size=64, shuffle=True)
loader_test = DataLoader(ds_test, batch_size=64, shuffle=False)
```

### 3. ë‹¤ì¸µ í¼ì…‰íŠ¸ë¡  

```python
from torch import nn
from torch import optim
```

784(28x28)ê°œì˜ ì…ë ¥ ë‰´ëŸ°, 100ê°œì˜ ì€ë‹‰ ë‰´ëŸ°, ê·¸ë¦¬ê³  10ê°œì˜ ì¶œë ¥ ë‰´ëŸ°(0~9)ì„ ê°€ì§€ëŠ” ë‹¤ì¸µ í¼ì…‰íŠ¸ë¡ ì„ êµ¬ì„±í•©ë‹ˆë‹¤.\
ReLU í™œì„±í™” í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ë¹„ì„ í˜•ì„±ì„ ì¶”ê°€í•˜ê³ , ìµœì¢… ì¶œë ¥ì€ 10ê°œì˜ í´ë˜ìŠ¤ì— ëŒ€í•œ ì ìˆ˜ë¥¼ ì œê³µí•©ë‹ˆë‹¤.

```python
model = nn.Sequential()
model.add_module('fc1', nn.Linear(28*28*1, 100))
model.add_module('relu1', nn.ReLU())
model.add_module('fc2', nn.Linear(100, 100))
model.add_module('relu2', nn.ReLU())
model.add_module('fc3', nn.Linear(100, 10))

print(model)
```

í¬ë¡œìŠ¤ ì—”íŠ¸ë¡œí”¼ ì†ì‹¤ í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ì˜ ì˜ˆì¸¡ê³¼ ì‹¤ì œ ë ˆì´ë¸” ì‚¬ì´ì˜ ì°¨ì´ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤. Adam ìµœì í™” ê¸°ë²•ì„ ì‚¬ìš©í•˜ì—¬ ëª¨ë¸ì˜ ê°€ì¤‘ì¹˜ë¥¼ ì—…ë°ì´íŠ¸í•©ë‹ˆë‹¤.

```python
# ì˜¤ì°¨í•¨ìˆ˜ ì„ íƒ
loss_fn = nn.CrossEntropyLoss()

# ê°€ì¤‘ì¹˜ë¥¼ í•™ìŠµí•˜ê¸° ìœ„í•œ ìµœì í™” ê¸°ë²• ì„ íƒ
optimizer = optim.Adam(model.parameters(), lr=0.01)
```

ë°ì´í„° ë¡œë”ì—ì„œ ë¯¸ë‹ˆë°°ì¹˜ë¥¼ í•˜ë‚˜ì”© êº¼ë‚´ ëª¨ë¸ì„ í›ˆë ¨í•©ë‹ˆë‹¤.\
 ì†ì‹¤ì„ ê³„ì‚°í•˜ê³  ì—­ì „íŒŒë¥¼ í†µí•´ ê¸°ìš¸ê¸°(gradient)ë¥¼ ê³„ì‚°í•œ ë‹¤ìŒ, ìµœì í™”ê¸°ë¥¼ ì‚¬ìš©í•˜ì—¬ ê°€ì¤‘ì¹˜ë¥¼ ì—…ë°ì´íŠ¸í•©ë‹ˆë‹¤.

```python
# ì´ 3ë²ˆì˜ ì—í¬í¬ ë™ì•ˆ ëª¨ë¸ í•™ìŠµ
epochs = 3

for epoch in range(epochs):
  for data, targets in loader_train:
    optimizer.zero_grad()      # ì˜µí‹°ë§ˆì´ì €ì˜ ê¸°ìš¸ê¸° ì´ˆê¸°í™”
    y_pred = model(data)          # ìˆœì „íŒŒ ì—°ì‚°ìœ¼ë¡œ ì˜ˆì¸¡ê°’ ê³„ì‚°
    loss = loss_fn(y_pred, targets)  # ì†ì‹¤ í•¨ìˆ˜ë¡œ ë¹„ìš© ê³„ì‚°
    loss.backward()            # ì—­ì „íŒŒ ì—°ì‚°ìœ¼ë¡œ ê¸°ìš¸ê¸° ê³„ì‚°
    optimizer.step()           # ì˜µí‹°ë§ˆì´ì €ë¥¼ í†µí•´ íŒŒë¼ë¯¸í„° ì—…ë°ì´íŠ¸

  print('Epoch {:4d}/{} Cost: {:.6f}'.format(epoch + 1, 3, loss.item()))
```

í…ŒìŠ¤íŠ¸ ë°ì´í„°ì— ëŒ€í•´ì„œ í‰ê°€ë¥¼ ì§„í–‰í•´ë´…ì‹œë‹¤. ëª¨ë¸ì„ í‰ê°€í•˜ê¸° ìœ„í•´ ì¶”ë¡  ëª¨ë“œë¡œ ì„¤ì •í•˜ê³ , ë°ì´í„° ë¡œë”ë¥¼ í†µí•´ í…ŒìŠ¤íŠ¸ ë°ì´í„°ì— ëŒ€í•œ ì˜ˆì¸¡ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤.\
ì˜ˆì¸¡ëœ í´ë˜ìŠ¤ì™€ ì‹¤ì œ í´ë˜ìŠ¤ë¥¼ ë¹„êµí•˜ì—¬ ì •í™•ë„ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.

```python
model.eval()  # ì‹ ê²½ë§ì„ ì¶”ë¡  ëª¨ë“œë¡œ ì „í™˜
correct = 0

# ë°ì´í„°ë¡œë”ì—ì„œ ë¯¸ë‹ˆë°°ì¹˜ë¥¼ í•˜ë‚˜ì”© êº¼ë‚´ ì¶”ë¡ ì„ ìˆ˜í–‰
with torch.no_grad():  # ì¶”ë¡  ê³¼ì •ì—ëŠ” ë¯¸ë¶„ì´ í•„ìš”ì—†ìŒ
    for data, targets in loader_test:

        outputs = model(data)  # ë°ì´í„°ë¥¼ ì…ë ¥í•˜ê³  ì¶œë ¥ì„ ê³„ì‚°

        # ì¶”ë¡  ê³„ì‚°
        _, predicted = torch.max(outputs.data, 1)  # í™•ë¥ ì´ ê°€ì¥ ë†’ì€ ë ˆì´ë¸”ì´ ë¬´ì—‡ì¸ì§€ ê³„ì‚°
        correct += predicted.eq(targets.data.view_as(predicted)).sum()  # ì •ë‹µê³¼ ì¼ì¹˜í•œ ê²½ìš° ì •ë‹µ ì¹´ìš´íŠ¸ë¥¼ ì¦ê°€

# ì •í™•ë„ ì¶œë ¥
data_num = len(loader_test.dataset)  # ë°ì´í„° ì´ ê±´ìˆ˜
print('\ní…ŒìŠ¤íŠ¸ ë°ì´í„°ì—ì„œ ì˜ˆì¸¡ ì •í™•ë„: {}/{} ({:.0f}%)\n'.format(correct, data_num, 100. * correct / data_num))
```

í‰ê°€ ì½”ë“œë¥¼ ì‚´í´ë´…ì‹œë‹¤. ì‹ ê²½ë§ ëª¨ë¸ì„ ì¶”ë¡  ëª¨ë“œë¡œ ì„¤ì •í•˜ì—¬ í…ŒìŠ¤íŠ¸ ë°ì´í„°ì— ëŒ€í•œ ì„±ëŠ¥ì„ í‰ê°€í•©ë‹ˆë‹¤. í•¨ìˆ˜ëŠ” model.eval()ì„ í˜¸ì¶œí•´ ëª¨ë¸ì„ ì¶”ë¡  ëª¨ë“œë¡œ ì „í™˜í•©ë‹ˆë‹¤. ì´ëŠ” ëª¨ë¸ ë‚´ì˜ ì¼ë¶€ ê³„ì¸µë“¤ì´ í•™ìŠµê³¼ ì¶”ë¡  ì‹œ ë‹¤ë¥´ê²Œ ë™ì‘í•  ìˆ˜ ìˆê¸° ë•Œë¬¸ì— í•„ìš”í•œ ì„¤ì •ì…ë‹ˆë‹¤.

torch.no_grad() ì»¨í…ìŠ¤íŠ¸ ë§¤ë‹ˆì €ëŠ” ì´ì–´ì§€ëŠ” ì½”ë“œ ë¸”ë¡ì—ì„œ ê¸°ìš¸ê¸° ê³„ì‚°ì„ ë¹„í™œì„±í™”í•˜ì—¬ ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰ì„ ì¤„ì´ê³  ê³„ì‚° ì†ë„ë¥¼ í–¥ìƒì‹œí‚µë‹ˆë‹¤. ì´ëŠ” í…ŒìŠ¤íŠ¸ ì‹œì—ëŠ” ëª¨ë¸ì„ ì—…ë°ì´íŠ¸í•˜ì§€ ì•Šê¸° ë•Œë¬¸ì— ê¸°ìš¸ê¸°ë¥¼ ì¶”ì í•  í•„ìš”ê°€ ì—†ê¸° ë•Œë¬¸ì…ë‹ˆë‹¤.

ë°ì´í„°ë¡œë” loader_testì—ì„œ ë¯¸ë‹ˆë°°ì¹˜ ë‹¨ìœ„ë¡œ ë°ì´í„°ì™€ ë ˆì´ë¸”ì„ ì¶”ì¶œí•˜ë©°, ëª¨ë¸ì€ ê° ë°ì´í„° ë°°ì¹˜ë¥¼ ë°›ì•„ í•´ë‹¹ ë°ì´í„°ì˜ ì¶œë ¥ outputsì„ ê³„ì‚°í•©ë‹ˆë‹¤. torch.max(outputs.data, 1)ëŠ” ê° ì˜ˆì¸¡ì— ëŒ€í•´ ê°€ì¥ ë†’ì€ í™•ë¥ ì„ ê°€ì§„ í´ë˜ìŠ¤ë¥¼ ì°¾ì•„ë‚´ê³ , ì´ëŠ” ëª¨ë¸ì´ ì˜ˆì¸¡í•œ ë ˆì´ë¸”ì´ ë©ë‹ˆë‹¤.

ì˜ˆì¸¡í•œ ë ˆì´ë¸” predictedì™€ ì‹¤ì œ ë ˆì´ë¸” targetsì„ ë¹„êµí•˜ì—¬ ì¼ì¹˜í•˜ëŠ” ê°œìˆ˜ë¥¼ ì„¸ì–´ correctì— ë”í•©ë‹ˆë‹¤. ë§ˆì§€ë§‰ìœ¼ë¡œ ì „ì²´ ë°ì´í„° ê°œìˆ˜ data_num ëŒ€ë¹„ ì •í™•í•˜ê²Œ ì˜ˆì¸¡í•œ ë°ì´í„°ì˜ ë¹„ìœ¨ì„ ê³„ì‚°í•˜ì—¬ í™”ë©´ì— ì¶œë ¥í•©ë‹ˆë‹¤. ì´ ë¹„ìœ¨ì€ ëª¨ë¸ì˜ ì •í™•ë„ë¥¼ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤

<br><br><br><br>
ì´ì œ ì„ì˜ì˜ ë°ì´í„°ì— ëŒ€í•´ì„œ íŠ¹ì • ì¸ë±ìŠ¤ì˜ í…ŒìŠ¤íŠ¸ ì´ë¯¸ì§€ë¥¼ ëª¨ë¸ì— ì „ë‹¬í•˜ì—¬ ì˜ˆì¸¡ ê²°ê³¼ë¥¼ ì–»ì–´ë´…ì‹œë‹¤. ê·¸ ê²°ê³¼ë¥¼ ì¶œë ¥í•˜ê³ , í•´ë‹¹ ì´ë¯¸ì§€ë¥¼ ì‹œê°í™”í•˜ì—¬ ì •ë‹µ ë ˆì´ë¸”ê³¼ í•¨ê»˜ ë³´ì—¬ì¤ë‹ˆë‹¤. ì´ë¥¼ í†µí•´ ëª¨ë¸ì˜ ì˜ˆì¸¡ì´ ì–¼ë§ˆë‚˜ ì •í™•í•œì§€ ì§ì ‘ í™•ì¸í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
index = 2018

model.eval()  # ì‹ ê²½ë§ì„ ì¶”ë¡  ëª¨ë“œë¡œ ì „í™˜
data = X_test[index]
output = model(data)  # ë°ì´í„°ë¥¼ ì…ë ¥í•˜ê³  ì¶œë ¥ì„ ê³„ì‚°
_, predicted = torch.max(output.data, 0)  # í™•ë¥ ì´ ê°€ì¥ ë†’ì€ ë ˆì´ë¸”ì´ ë¬´ì—‡ì¸ì§€ ê³„ì‚°

print("ì˜ˆì¸¡ ê²°ê³¼ : {}".format(predicted))

X_test_show = (X_test[index]).numpy()
plt.imshow(X_test_show.reshape(28, 28), cmap='gray')
print("ì´ ì´ë¯¸ì§€ ë°ì´í„°ì˜ ì •ë‹µ ë ˆì´ë¸”ì€ {:.0f}ì…ë‹ˆë‹¤".format(y_test[index]))
```
