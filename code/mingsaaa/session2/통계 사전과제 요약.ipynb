## 1. 통계란?

### 1-1) 들어가기 전에

통계학은 이 두 가지 측면으로 나뉩니다:  
- **기술통계**: 수치와 시각화를 통해 샘플을 요약하는 방법론
- **추론통계**: 추정과 검정을 통해 샘플로부터 모집단을 추론하는 방법론

### 1-2) 왜 통계를 공부해야 하나요?

- 통계학은 데이터를 수집하고 분석하여 패턴과 관계를 찾아내고 결론을 도출하는 학문입니다.
- 본질적으로, 통계학은 데이터로부터 의미를 찾는 것입니다.
- 데이터 분석의 궁극적인 목적은 **미래 예측**입니다.
- 정보기술의 발전으로 데이터는 쉽게 구할 수 있지만, 의미 있는 정보로 변환하기 위해서는 통계학적 지식이 필요합니다.

### 1-3) 데이터의 척도

- **양적 자료**(quantitative, numerical data): 숫자로 표현할 수 있는 데이터
- **질적 자료**(qualitative, categorical data): 숫자로 표현할 수 없는 데이터

양적 자료는 **구간척도**와 **비율척도**로 구분되고, 질적 자료는 **명목척도**와 **서열척도**로 구분됩니다.

#### 1. 명목척도 (nominal scale)
- 단순한 범주만 나타내는 척도. 숫자는 분류의 기능만 하며, 양적인 의미는 없음.
- 예: 색깔 (빨 : 0, 주 : 1, 노 : 2), 성별 (남 : 0, 여 : 1), 취미 (운동 : 0, 독서 : 1)

#### 2. 서열척도 (ordinal scale)
- 순위 또는 서열이 있는 척도로 수량화할 수 있음. 하지만 차이를 말할 수는 없음.
- 예: 달리기 순위 (1등, 2등), 학년 (1학년, 2학년), 만족도 (낮음 : 0, 보통 : 1, 높음 : 2)

#### 3. 구간(등간)척도 (interval scale)
- 연속적인 수로 수량화할 수 있으며, 숫자간의 간격이 동일함. 절대적인 원점(0)은 존재하지 않음.
- 덧셈과 뺄셈은 가능, 곱셈과 나눗셈은 불가능.
- 예: 온도, IQ

#### 4. 비율척도 (ratio scale)
- 연속적인 수로 수량화할 수 있으며, 절대적인 원점(0)이 존재함.
- 모든 사칙연산이 가능.
- 예: 몸무게, 인구수, 길이

## 2. (수치) 기술 통계

### 2-1) 기술통계란?

**기술 통계학 (Descriptive Statistics)**  
- 데이터의 주요 특성을 요약하고 설명하는 방법을 다루는 분야입니다. 평균, 중앙값, 표준편차 등이 이에 해당합니다.  
- **수치 기술 통계**: 데이터를 하나의 수치로 요약하고 설명하는 방법 (중심 경향성, 산포도 등)  
- **시각화 기술 통계**: 표와 그래프를 사용하여 데이터를 시각화하는 방법  

### 2-2) 중심 경향성

- **평균 (mean)**: 데이터 값의 총합을 데이터의 개수로 나눈 값으로, 데이터의 중심 경향성을 나타내는 지표입니다.  
    예: `2, 2, 3, 5, 8` → 평균 = (2 + 2 + 3 + 5 + 8) / 5 = 4  
- **중앙값 (median)**: 데이터 값을 크기 순서대로 나열했을 때 가운데 값입니다.  
    예: `2, 2, 3, 5, 8` → 중앙값 = 3  
- **최빈값 (mode)**: 데이터에서 가장 자주 나타나는 값입니다.  
    예: `2, 2, 3, 5, 8` → 최빈값 = 2  
- **왜도 (skewness)**: 데이터 분포의 비대칭성을 나타내는 지표입니다.  
    - **Positive Skewness (Right-skewed)**: 오른쪽 꼬리가 더 긴 분포  
    - **Negative Skewness (Left-skewed)**: 왼쪽 꼬리가 더 긴 분포  
    - **Normal distribution**: 대칭적인 정규분포

### 2-3) 산포도

- **분산 (variance)**: 데이터 값들이 평균으로부터 얼마나 멀리 퍼져있는지를 나타냅니다.  
    np.var(arr)
- **표준편차 (standard deviation)**: 분산에 제곱근을 취한 값으로, 데이터의 변동성을 나타냅니다.  
    np.std(arr)
- **범위 (range)**: 데이터의 최대값과 최소값의 차이입니다.  
    np.max(arr) - np.min(arr)
- **사분위수 (quartile)**: 데이터를 네 개의 동일한 크기로 나눈 값입니다.  
    - **Q1**: 하위 25%  
    - **Q2**: 하위 50% (중앙값)  
    - **Q3**: 하위 75%  
    np.percentile(arr, 25)`, `np.percentile(arr, 50)`, `np.percentile(arr, 75)
- **사분위간 범위 (IQR)**: Q3에서 Q1을 뺀 값으로, 데이터의 중앙 50% 범위를 나타냅니다.  
    - **아래쪽 이상치**: Q1 − 1.5×IQR 보다 작은 값  
    - **위쪽 이상치**: Q3 + 1.5×IQR 보다 큰 값
- **변동계수 (coefficient of variation)**: 표준편차를 평균으로 나누어 상대적인 변동성을 나타냅니다.  
    np.std(arr) / np.mean(arr) * 100

## 3. **추론 통계 - 추정**

### 3-1) 추론통계란?

**추론 통계학 (Inferential Statistics)**  
- 수집한 데이터를 기반으로 미래의, 혹은 조사하지 못한 전체 정보를 추론하고 결론을 도출하는 방법을 다루는 분야입니다.  
- **통계적 가설검정**과 **신뢰구간** 등이 포함됩니다.  
- 모집단에서 일부를 뽑아 만든 표본 집단을 통해 모집단에 대한 정보를 추측하는 과정입니다.

### 3-2) 확률분포

- **확률분포** (probability distribution): 확률변수가 특정한 값을 가질 확률을 나타내는 함수입니다.  
- **확률변수**: 어떤 특정 사건의 발생을 수치로 대응시키는 함수입니다. 예: 동전 던지기 (앞면=1, 뒷면=0)  

**확률분포의 종류**  
- **이산확률분포**: 확률 변수가 이산적인 값을 가질 때  
- **연속확률분포**: 확률 변수가 연속적인 값을 가질 때

### 3-3) 대표적인 분포 모형

- **정규분포 (Normal Distribution)**  
    - **형태**: 종 모양 (bell curve), 평균을 기준으로 대칭  
    - **매개변수**: 평균(μ), 표준편차(σ)  
    - **특징**: 자연 현상과 사회 현상에서 많이 나타남, 중심극한정리에 의해 표본 크기가 커지면 정규분포에 근사  
    - **대수의 법칙**: 표본 크기가 커질수록 표본 평균이 모집단의 평균에 근사  
    - **중심극한정리**: 표본 크기가 커질수록 표본 평균의 분포는 정규분포에 가까워짐

- **T분포 (T-Distribution)**  
    - **형태**: 종 모양, 정규분포보다 꼬리가 두꺼움  
    - **매개변수**: 자유도(df)  
    - **특징**: 표본 크기가 작을 때 사용, 표본의 분산을 통해 모집단 분산을 추정

- **카이제곱분포 (Chi-Square Distribution)**  
    - **형태**: 오른쪽으로 긴 꼬리를 가진 비대칭 분포  
    - **매개변수**: 자유도(df)  
    - **특징**: 표준 정규 분포를 따르는 변수들의 제곱 합, 독립성 검정, 적합도 검정 등에 사용


## 4. 가설 검정 (Hypothesis Testing)

가설 검정은 모집단의 실제 값에 대한 주장과 관련해, 표본 정보를 사용하여 가설의 합당성 여부를 판별하는 과정입니다.
## 4-1) 가설 검정의 절차
### 1. 가설 설정
- **귀무가설 (H0)**: 기각하고자 하는 가설, 보통 "~는 차이가 없다"라는 명제.
- **대립가설 (H1)**: 주장하고자 하는 가설, 보통 "~는 차이가 있다"라는 명제.

#### 예시:
- 귀무가설: "마포구와 서대문구의 인구 평균 키는 같다"
- 대립가설: "마포구와 서대문구의 인구 평균 키는 다르다"

### 2. 유의 수준 설정
- **유의 수준 (α)**: 연구자가 허용할 수 있는 오류의 최대치. 보통 0.05(5%)로 설정합니다.
- **1종 오류 (𝛼)**: 귀무가설이 참일 때 기각할 확률.
- **2종 오류 (𝛽)**: 귀무가설이 거짓일 때 기각하지 못할 확률.

### 3. 검정 통계량 산출
- **Z-Score (Z값)**: 표준 정규 분포에서 사용하며, 표본 크기가 크거나 모집단의 분포가 정규일 때 사용됩니다.
    - 공식: $Z = \frac{X - \mu}{\sigma}$
    - 임계값: Z = 1.96 (유의수준 0.05)
    
- **T-Score (T값)**: 표본이 작은 경우에 사용되며, T분포에서 기초합니다.
    - 공식: $T = \frac{X - \mu}{s / \sqrt{n}}$
    - T분포에서 자유도에 따라 임계값이 달라집니다.

- **F-Score (F값)**: 두 표본 분산 간의 차이를 비교할 때 사용됩니다.
    - 공식: $F = \frac{\text{분산}_1}{\text{분산}_2}$

### 4. 가설 기각/채택 판단
- **유의확률 (p-value)**: 검정통계량에서 유도할 수 있는 값으로, p-value가 유의수준(α)보다 작으면 귀무가설을 기각합니다.
    - 만약 p-value < α, 귀무가설 기각
    - p-value > α, 귀무가설 기각하지 않음
