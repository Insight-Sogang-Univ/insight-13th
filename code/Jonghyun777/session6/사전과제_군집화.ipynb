 ### **1. 군집화**

***(1)머신러닝:비지도학습***

군집화는 비지도 학습에 속함. 비지도 학습은 패턴이 알려지지 않았거나 계속해서 변화하는 문제를 해결할 때 유용.
구체적이고 좁은 범위에서 정의된 문제를 푸는 데도 지도학습보다 불리하지만, 더 열린 문제를 해결하고 지식을 일반화하는 데는 유리함

***(2) 군집화란?***

군집화란 데이터를 비슷한 특성을 가진 그룹으로 나누는 비지도 학습 기법  

-> 각 그룹이 어떻게 구성되어있는지 파악하면서 전체 데이터 구조를 이해할 수 있음.

*군집*: 유사한 데이터들의 집합. 이때 서로 다른 군집에 속한 개체들은 비슷하지 않아야 한다. 

*군집화의 목표*  
> - 응집도 최대화: 같은 군집에 속하는 데이터끼리는 최대한 비슷하도록 함
> - 분리도 최대화: 서로 다른 군집은 최대한 분리되도록 함.

<BR>-> 군집화를 통해 데이터 간의 유사성을 최대한 잘 유지하면서, 서로 다른 그룹은 구분될 수 있도록 만드는 것이 군집화의 목표

***(3) 군집화 과정***

1. 피처 선택 또는 추출  
2. 군집화 알고리즘 선택  
3. 군집 유효성 검증  
4. 결과 해석 

*군집화에서 주요하게 고려해야하는 사항*

1. 변수 유형 이해: 변수 종류가 무엇인지.변수 개와 특징이 무엇인지 이해 필요. 변수가 종류와 특징에 따라서 사용하는 방법론을 결정하기 떄문

2. 거리(또는 유사도) 정의와 측정 : 군집화는 데이터 간의 유사도를 기반으로 이루어지기 때문에,어떤 방식으로 거리 또는 유사도를 측정하는지가 중요

3. 차원 축소: 유사한 변수들을 묶어서 차원을 축소  
 ex) 연봉,나이. 비슷하기에 하나만을 사용

### **2. 군집화 알고리즘**

***(1) 계층적 군집화***

: 데이터 간의 유사성을 기반으로 트리 구조(dendrogram)를 형성하며, 상향식 또는 하향식 방식으로 군집을 형성해 나가는 방법

상향식 트리구조-> 응집형 계층적 군집화(agglomerative hierarchical clustering) or 병합 클러스터링  
하향식 트리구조-> 분리형 계층적 군집화(divisive hierarchical clustering) or 분할 클러스터링  
*작은 군집들을 병합해가며 더 큰 군집을 만드느냐, 하나의 군집을 분할해가며 여러 군집을 만드느냐의 차이*

> - 병합 클러스터링 : 먼저 각 샘플이 독립적인 클러스터가 됨. 이후 하나의 클러스터만 남을 때까지 가장 가까운 클러스터를 병합
> - 분할 클러스터링: 전체 샘플을 포함하는 하나의 클러스터에서 시작하여, 유사성이 낮은 데이터들을 더 작은 클러스터로 나눔

**응집형 계층적 군집화에 집중 레츠고**

응집형 계층 군집화의 거리 정의 알고리즘은 단일 연결과 완전 연결
> - 단일 연결: 클러스터 쌍에서 가장 비슷한 샘플 간 거리를 계산, 이후 이 거리가 가장 작은 두 클러스터를 병합
> - 완전 연결: 단일 연결과 비슷하지만, 가장 비슷한 샘플이 아니라 가장 비슷하지 않은 샘플을 비교해 볍합을 수행

[이외의 6가지 정의 방법]

1. single linkage(단일 연결법/최다 연결법)


: 서로 다른 군집에서 가장 가까운 두 점 사이의 거리를 군집 간의 거리로 측정하는 방법. 고립된 군집을 찾는데 중점을 두고 있으며 이상치에 취약

2. Complete lingkage(완전 연결법.최강 연결법)

: 군집에서 하나씩 관측값을 뽑았을 때 나타날 수 있는 거리의 최댓값을 군집에서 하나씩 관측값을 뽑았을 때 나타날 수 있는 거리의 최댓값으로 측정하여 가장 유사성이 큰 군집으로 병합해 나가는 방법으로, 군집들의 내부 응집성에 중점을 둔 방법. 최단 연결법과 마찬가지로 이상치에 민감함

3. Average linkage( 평균 연결법)

: 모든 항목에 대한 거리 평균을 구함. 계산량이 불필요하게 많아질 수 있다는 단점이 있으나, 최단 연결법과 최장 연결법보다 이상치에 덜 민감함.

4. Centroid Method(중심 연결법)

:두 군집의 중심 간의 거리를 측정. 두 군집이 결합될 때 새로운 군집의 평균은 가중평균을 통해 구함. 군집들의 중심을 계산해야 하기 때문에 시간이 오래 걸린다는 

5. Median (중간 연결법)

: 군집간의 거리를 군집 내 모든 샘플의 중앙값으로 정의. 데이터의 중앙값에 기반해 군집 간 거리를 정의하므로 극단값에 영향을 덜 받으나, 기하학적 구조를 파악하기는 어렵다는 특징이 있음

6. Ward's prodcedure(Ward 연결법)

: 군집의 병합 후 군집 내 SSE의 증가분 최소인 것을 선택. 보통 두 군집이 합해지면 병합된 군집의 SSE는 병합 이전 각 군집의 SSE의 합보다 커지게 되는데 그 증가량이 가장 작아지는 방향으로 군집을 형성해 나가는 방법

`완전 연결 방식을 사용한 응집형 계층 군집화 과정`

1. 모든 샘플의 거리 행렬을 계산
2. 모든 데이터 포인트를 단일 클러스터로 표현
3. 가장 비슷하지 않은 샘플 사이 거리에 기초하여 가장 가까운 두 클러스터를 합침
4. 유사도 행렬을 업데이트
5. 하나의 클러스터가 남을 때까지 2~4번을 반복

***(2) K means***

: 데이터를 정해진 개수(k)의 그룹으로 나누되, 각 그룹의 중심점과의 거리가 가장 가까운 데이터끼리 묶는 군집화 알고리즘

[K-means의 장점]
> - 직관적이고 구현이 쉬움
> - 대용량 데이터에도 적용이 가능

[K-means의 단점]

> - 초기 centroid 값에 민감함
> - 군집 수(k) 결정이 어려움
> - 아웃라이어에 민감
> - 기하학적인 모양의 군집은 파악하기 어려움

[k-means ++ 알고리즘]

k-means에서는 초기 중심점을 랜덤하게 할당. 이 떄문에 초기 중심점에 따라 최종군집화의 품질이 달라지거나 알고리즘 성능 하락.  
 이를 해결하기 위해 k-means 알고리즘을 여러번 실행해 SSE 입장에서 가장 성능이 좋은 모델을 선택.  
 또 다른 방법으로는 k-means ++ 알고리즘을 통해 초기 중심점들을 서로 멀리 떨어진 곳에 위치시키는 것이 있음. 이를 통해 기본적인 k-means 보다 일관되고 좋은 결과를 도출 가능



[옐보우 방법(elbow method)]

k-means의 k값에 따라 군집 품질이 달라진다는 단점을 해결하기위해 클래스 내 SSE를 바탕으로 엘보우 방법이라고 하는 그래프를 활용해 최적 클러스터 개수 k를 추정 가능  
직관적으로 생각했을 때, k가 증가하면 왜곡은 줄어들 것임->샘플이 할당된 중심점에 더 가까워지기 떄문

***(3)DBSCAN***

: 밀도가 높은 지역의 데이터를 하나의 군집으로 묶고, 밀도 기준을 만족하지 못하는 점은 군집에 포함시키지 않는 군집화 알고리즘

[DBSCAN 알고리즘의 주요 단계]


> - 1) 어떤 샘플의 특정 반경  $\epsilon$ 안에 있는 이웃 샘플이 지정된 개수 이상이면 핵심 샘플(core point)이 됨
> - 2)  $\epsilon$ 이내에 MinPts보다 이웃이 적지만 다른 핵심 샘플의 반경  $\epsilon$안에 있으면 경계 샘플(border point)이 됨
> - 3) 위 두 샘플에 해당하지 않는 다른 모든 샘플은 잡음 샘플(noise point)이 됨





2. 개별 핵심 샘플 또는( $\epsilon$ 이내에 있는 핵심 샘플을 연결한) 핵심 샘플의 그룹을 클러스터로 만들어줌
3. 경계 샘플을 해당 핵심 샘플의 클러스터에 할당

[DBSCAN의 장단점]


DBSCAN의 장점
- k-means처럼 클러스터 모양을 원형으로 가정하지 않고, 밀도에 따라 클러스터를 할당. → 임의의 기하학적 분포를 갖는 데이터도 잘 처리할 수 있음.
- 이상치에 민감x : Noise를 통해 이상치 검출 가능
- k-means나 계층 군집과 달리 모든 샘플을 클러스터에 할당하지 않음

DBSCAN의 단점
- `epsilon`과 `min_samples` 설정에 많은 영향을 받음
- `epsilon`을 조절하기 쉽지 않으며, 적절한 조절을 위해서는 데이터셋에 대한 도메인 지식이 필요한 경우가 많음
- 연산량이 많아 속도가 느림
- 다른 밀도 분포를 가진 데이터의 군집 분석을 잘 하지 못함


->밀도가 높은 곳에 집중하다 보니 왼쪽에 밀도가 낮은 곳에 데이터를 하나의 군집으로 인식하지 못하고 noise point로 구분해버림
- 차원의 저주를 벗어나지못함: 모든 유클리디안 거리를 계산하는 알고리즘이 갖는 문제 

(4) Gaussian Mixture Model(GMM)

: 데이터가 여러 다른 모양의 가우시안 분포로 구성되었다고 가정하고, 각 분포를 클러스터로 인식하는 군집화 방법

GMM은
> - 데이터가 해당 모델로부터 생성되었다는 전제 하에 수행하는 접근 방식
> - 각 군집을 확률 분포로 간주
> - 전체 데이터 분포를 이질적인 여러 개의 확률 분포의 혼합으로 보고 모델링
> - 데이터의 생성 메커니즘을 모델링

[GMM 진행과정]

[GMM의 장단점]

- **GMM의 장점**
    - K-means보다 유연하게 다양한 데이터 세트에 잘 적용될 수 있음 : k-means가 잘 작동하지 않는 타원형 분포나, 중첩된 군집 구조에서도 좋은 성능을 낼 수 있다.
- **GMM의 단점**
    - 군집화를 위한 수행시간이 오래 걸림.
    - 가정한 분포(가우시안 분포)에 맞지 않은 데이터일 경우 계산 복잡도가 높아지고 성능이 저하될 수 있음.

 ### **3.군집화의 평가 방법**

비지도 학습에서 가장 어려운 점으 답을 모든다는 점. 군집화에서는 군집 품질을 평가하기 위해 알고리즘 자체의 **내부 평가 지표**를 사용해야 함. 예를들어 k_means같은 경우 클래스 내 SSE(왜곡) 사용

내부 평가는 군집 내부의 응집도와 군집 간의 분리도를 정량적으로 측정하여 데이터 자체의 구조를 기반으로 군집 품질을 판단  
정답 레이블이 존재하는 경우에도 비지도 학습 방법을 적요해볼 수 잇는데 이때는 **외부 평가 지표**를 활용해 군집화 결과를 평가할 수 있음.


1. 내부 평가 (Internal Evaluation)
    - 정답 라벨이 없는 경우 : 군집 내부의 응집도와 군집 간의 분리도를 기준으로 군집 품질을 평가합니다.
    - 대표적인 지표
        - 실루엣 계수(Silhouette Coefficient) : 한 점이 자기 클러스터 내에서 얼마나 가깝고, 다른 클러스터와는 얼마나 멀리 떨어져 있는지 측정
2. 외부 평가 (External Evaluation)
    - 정답 라벨이 있는 경우, 실제 라벨과 군집화 결과를 비교할 수 있습니다.
    - 참고 - 대표적인 지표 : **ARI** (Adjusted Rand Index)
        - 두 군집 결과 간의 유사도를 비교하는 지표인 **Rand Index(RI)**를 조정하여 한계를 보완한 지표가 **ARI**입니다.
          

(1) 실루엣 계수

: 클러스터 내 샘플들이 얼마나 조밀하게 모여 있는지를 측정하는 도구. 이를 통해 군집의 품질을 확인

[실루엣 계수 계산 과정]

개별 샘플에 대한 **실루엣 계수(silhouette coefficient)**를 계산하는 과정

1. 샘플 $\mathbf{x}^{(i)}$와 동일한 클러스터 내 모든 다른 포인트 사이의 거리를 평균하여 **클러스터 응집력(cluster cohesion)** $\mathbf{a}^{(i)}$ 를 계산
2. 샘플 $\mathbf{x}^{(i)}$와 가장 가까운 클러스터의 모든 샘플 간 평균 거리로 최근접 클러스터의 **클러스터 분리도 (custer separation)** $\mathbf{b}^{(i)}$ 를 계산합니다.
3. 클러스터 응집력과 분리도 사이의 차이를 둘 중 큰 값으로 나누어 실루엣 $s^{(i)}$ 를 다음과 같이 계산합니다.
    
    $$
    s^{(i)} = \frac{b^{(i)} - a^{(i)}}{\max \left\{ b^{(i)}, a^{(i)} \right\}}
    $$

(2) Dunn Index

 : 클르서터 간 최소거리와 클러스터 내 최대거리의 비율을 계산해 클러스터링의 품질을 평가하는 지표

$$
\text{Dunn Index} = \frac{
\min\limits_{i \ne j} \; d(C_i, C_j)}{\max\limits_{1 \le k \le K} \; \delta(C_k)}
$$

Dunn index 값이 클수록 좋은 군집화 결과를 의미. 군집화는 군집 간 거리가 크고(분리도가 높음), 군집 내부 거리는 작을수록(응집도가 높음) 좋기 때문

[Dunn Index의 한계]
> - 군집 수가 많아질수록 계산 비용이 증가
> - 이상치에 민감할 수 있음
