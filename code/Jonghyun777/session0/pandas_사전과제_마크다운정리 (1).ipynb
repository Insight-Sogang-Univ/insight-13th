### **2. 데이터프레임의 구성**

###### 데이터프레임이란 컴퓨터로 정보를 정리하고 저장하는 도구. 판다스에서 사용하는 주요 데이터 구조


#### 2-1) Index와 Column


데이터 프레임의 기본 구조는 **행과 열**로 설명 
> - 가로-> 행 세로-> 열
> - 행과 열은 **데이터의 위치를 나타내는 역할**
> - 인덱스-> 데이터의 순서 , 컬럼-> 데이터의 카테고리

### 2-2) Series 

> - 시리즈는 값(value)과 인덱스(Index)로 구성 1차원 자료구조
> - 인덱스가 있는 1차원 배열(array)로 볼 수 있고, 시리즈를 구성하는 값들의 데이터 타입은 모두 같아야함
> - 별도 인덱스 레이블 지정 x -> 0부터 시작되는 디폴트 정수형 인덱스를 자동으로 사용
> - 데이터프레임에서 특정 컬럼을 1개 만 추출한다면 시리즈라고 볼 수 없음

**시리즈 만드는 법**


1. 원하는 값들이 포함된 리스트 생성
```python
series = ['LG', '애플', '애플', '삼성', 'ASUS', '삼성']


2. pd.Series() 함수를 사용하여 시리즈 개체 생성
```python
sr = pd.Series(series)


인덱스를 직접 지정해주고 싶으면 pd.Series()함수의 인자에 인덱스를 추가
```python
index = [1,2,3,4,5,6]
sr = pd.Series(series, index = index)


#### 2-3) DataFrame


> - 데이터 프레임은 행과 열로 이루어진 **2차원 배열 구조**
> - 길이가 같은 시리즈들의 집합체
> - 시리즈, 즉 칼럼 단위로는 값들의 데이터 타입이 동일해야 하지만, 칼럼 간의 데이터 타입은 달라도 무방
     

**데이터프레임 만드는 법**

1. 데이터와 각 카테고리 별 네이밍을 칼럼으로 하는 딕셔너리 생성

```python
data = {
    '노트북 브랜드' : ['LG', '애플', '애플', '삼성', 'ASUS', '삼성'],
    '노트북 화면 크기(인치)' : [14.0, 14.2, 15.3, 13.3, 14.0, 15.6],
    '노트북 무게(kg)' : [0.99, 1.63, 1.51, 0.87, 1.28, 1.80],
    '사용자 이름' : ['연동운', '문유정', '이다빈', '최재완', '조영진', '김진형'],
    
		

2.pd.DataFrame() 함수를 사용하여 데이터프레임 구성

```python
df = pd.DataFrame(data)

인덱스, 컬럼을 직접 지정하는 것도 가능 (기본값은 둘다 0,1,2…)

```python
index = [1,2,3,4,5,6]
columns = ['브랜드', '화면 크기', '무게', '이름']
df = pd.DataFrame(data, index=index, columns=columns)

#### 2-4) 데이터 확인하기


`본격적인 데이터 분석을 위해선 데이터프레임에 대한 기본적인 이해 필요
우리가 마주하게 될 데이터는 크기가 매우 방대할 것.` 

따라서 우리는 데이터의 일부를 확인하거나 전체적인 경향을 볼 수 있는 함수를 통해 데이터를 봐야함

 기본적인 데이터 확인 방법  
 _(데이터를 분석할 때 가장 먼저 해야 할 작업들!!)_  

**데이터프레임의 앞, 뒤 일부만 보기**  
- `df.head(n)` : 앞 `n`개의 행 반환  
- `df.tail(n)` : 뒤 `n`개의 행 반환  


**데이터에 대한 정보 확인하기**  
- `df.shape` : _(행 개수, 열 개수)_ 반환  
- `df.dtypes` : 열들의 데이터 타입 반환  
- `df.describe()` : 수치형 열들의 기본적인 통계값 반환  
- `df["열 이름"].unique()` : 특정 열에 포함된 값을 **종류별로 개수만 반환**  
- `df.isnull()` : 데이터에 **결측치(NaN)** 가 있는지 여부 반환  
- `df.info()` : 여러 정보를 종합적으로 반환  

**데이터의 개수 세기**  
- `df.count()` : 결측치를 제외한 데이터의 행 수 반환  
- `df["열 이름"].value_counts()` : 특정 열에 포함된 **값들의 빈도수**를 세어 반환  

### **3. 데이터 추출**

`우리가 분석하게 될 데이터는 양이 매우 방대함. 효율적인 분석을 위해서는 데이터를 가공하는 스킬이 중요.
때문에 데이터 추출을 해야함`

#### 3-1) 인덱싱 & 슬라이싱

 인덱스와 칼럼을 사용해 데이터를 추출하는 것을 인덱싱 & 슬라이싱 이라 함
 > - 인덱싱: 입력하는 조건이 범위가 아니라 단순 지목일 경우
>  - 슬라이싱: 범위를 조건으로 할 경우
    > - `df[start:end]`
```python
# 범위 슬라이싱
data[:1] # 0번째 데이터 반환
data[1:3] # 1~2번째 데이터 반환
data[2:] # 2번째~끝까지 데이터 반환


#### 3-2) 행단위 데이터 추출: loc과 iloc



 ** loc (라벨 기반 인덱싱)**
> - 데이터를 **행 단위로 추출**할 때 사용하는 메서드.
> - **형식:** `df.loc[행 이름, 열 이름]`
> -**인덱싱 & 슬라이싱 모두 가능**하며, **열 단위 혹은 특정 값만 추출**할 수도 있음.
> - `loc` 안에는 반드시 **"행과 열의 이름(label)"**이 들어가야 함.

 Ex)
```python
df.loc["행 이름"]         # 행 단위 추출
df.loc["행 이름", "열 이름"]  # 값 단위 추출
df.loc[:, "열 이름"]      # 열 단위 추출

** iloc (정수형 인덱스 기반)**
> - loc과 유사하지만 행과 열의 이름이 아닌 정수형 인덱스 번호 사용
> - 데이터가 크거나 속도가 중요한 경우 iloc이 더 빠름
> - 형식: `df.iloc(행 순서, 열 순서)`

Ex)
```python
df.iloc[0]          # 첫 번째 행 추출
df.iloc[0, 1]       # 첫 번째 행, 두 번째 열의 값 추출
df.iloc[:, 2]       # 세 번째 열 전체 추출
df.iloc[1:3, :]     # 2번째~3번째 행 전체 추출



#### 3-3) 열단위 데이터 추출: 컬럼명

열 단위로만 데이터를 추출하고 싶으면 **열 이름으로 인덱싱**하는 것이 더 좋은 방식

#### 3-4) 불리언 인덱싱

`인덱스나 컬럼명이 아닌 특정 조건에 만족하는 데이터를 추출하고 싶을 때 사용`

> - 불리언 인덱싱이란 True 또는 False의 배열을 통해 데이터를 필터링하는 연산을 말함
> - 데이터 추출을 위해 인덱스/칼럼명을 넣었던 자리에 조건문이 들어가게 되면 **True 값에 해당 하는 부분만 출력**
> - 조건문 여러 개를 논리연산자로 묶어 사용하는 것도 가능
> - loc에서도 불리언 인덱싱 사용 가능 

#### 3-5) 인덱스 설정

인덱스는 0부터 시작하는 정수형 인덱스를 기본값으로 가지지만, 원하는 대로 변경할 수 있음

### **4. 데이터 조작**

#### 4-1) 데이터 추가, 삭제 

> - 특정 데이터 셋을 열로 추가 가능
> - 열 삭제 가능
>   

#### 4-2) 데이터 정렬 

> - 데이터를 새로운 방식으로 정렬할 수 있음


#### 4-3) 데이터 변환

> - 특정 컬럼 내 데이터의 자료형을 변환 가능
> - 데이터의 값 자체를 변환할 수도 있음

#### 4-4) 데이터의 프레임 병합

> - 행방향,열방향 결합을 통해 두 데이터 프레임을 이어붙일 수 있음
> - 공통 항목을 기준으로 병합도 가능
> - on 인자를 통해 기준이 될 열을 지정하는 것도 가능

### **5. 집계함수**

#### 5-1) 집계함수란?

`여러 행으로부터 하나의 결괏값을 변환하는 함수`

평균,중앙값,표준편차,합계,최댓값,최소값이 대표적인 집계함수

#### 5-2) 그룹화(Grouping)

`집계함수는 기본적으로 모든 행을 대상으로 연산을 진행하지마 일부 대상으로 집계할 떄 더 유용할 수 있음`

> - **groupby** 메소드는 특정 기준으로 그룹을 묶고,연산 처리를 할 수 있게 해줌
> - 보통 집계함수와 함께 시용
