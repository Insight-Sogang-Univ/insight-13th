# Pandas

## 1. 데이터 프레임 구성
Pandas의 **DataFrame**은 2차원 **테이블** 형태로, **행**(index)과 **열**(columns)로 이루어져 있다.

### 데이터 프레임의 기본 구조:
- **행(row)**: 데이터의 순서 표시.
- **열(column)**: 데이터의 카테고리 표시, 동일한 타입의 데이터를 포함.
- **Series**: ‘값(value)’과 ‘인덱스(index)’로 구성된, 가장 간단한 1차원 자료구조. 
- **DataFrame**: 행과 열로 이루어진 2차원 배열 구조, 길이가 같은 시리즈의 집합체.
  
### 데이터프레임 만드는 법
1. 데이터와 각 카테고리 별 네이밍을 컬럼으로 하는 딕셔너리 생성
2. pd.DataFrame() 함수를 사용하여 데이터프레임 구성 (인덱스, 컬럼을 직접 지정하는 것도 가능)

## 데이터 확인하기
- **앞, 뒤 일부만 보기**: df.head(n), df.tail(n)
- **정보 확인**: df.shape, df.dtypes, df.describe(), df[열 이름].unique(), df.isnull(), df.info()
- **개수 세기**: df.count(), df[열 이름].value_counts()

  
## 2. 데이터 추출

### 인덱싱&슬라이싱
인덱스와 컬럼을 사용해 데이터를 추출하는 것을 ‘인덱싱 & 슬라이싱’이라 한다.
- 인덱싱 : 입력하는 조건이 범위가 아니라 **단순 지목**일 경우
- 슬라이싱 : **범위**를 조건으로 할 경우 ex) df[start:end]

### 행단위 데이터 추출: loc&iloc
- loc : 기본적으로 행단위 데이터를 추출 ex) df.loc[행 이름], df.loc[행 이름, 열 이름], df.loc[:, 열 이름]
- iloc : 정수형 인덱스 번호가 와야함 ex) df.iloc[행 순서, 열 순서]

### 열단위 데이터 추출
열 이름으로 인덱싱
ex) df[열 이름], df[[열 이름1, 열 이름2...]], df[[열 이름]], df[열 이름][행 이름]

### 불리언 인덱싱
True 또는 False의 배열을 통해 데이터를 필터링
ex) df[조건문], , df[(조건문1) 논리연산자 (조건문2)], df.loc[조건문, 열 이름]

### 인덱스 설정
df.set_index(열 이름), df.reset_index(), df.reindex(인덱스 배열)로 인덱스를 변경할 수 있다. 


## 3. 데이터 조작
### 데이터 추가 & 삭제
-**추가**: df['새로운 열 이름'] = 리스트
_**삭제**: df.drop(labels, axis)

### 데이터 정렬
-df.sort_index: 인덱스 기준  정렬
-df.sort_values["열 이름"]: 특정 열의 값 기준 정렬

### 데이터 변환
-특정 칼럼 내 **자료형** 변환: df[열 이름] = df[열 이름].astype(바꿀 자료형)
-**데이터 값** 변환: df.replace(바꾸고 싶은 값, 바꿀 값)

### 데이터프레임 병합
- concat: 두 데이터프레임의 **단순병합**
- merge: **공통항목**을 기준으로 병합
